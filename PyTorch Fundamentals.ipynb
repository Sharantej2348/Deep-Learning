### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning

1. Element-wise multiplication
2. Matrix multipliaction(dot product)

The main two rules for matrix multiplication to remember are:

1. The inner dimensions must match:
* (3, 2) @ (3, 2) won't work
* (2, 3) @ (3, 2) will work
* (3, 2) @ (2, 3) will work
2. The resulting matrix has the shape of the outer dimensions:
* (2, 3) @ (3, 2) -> (2, 2)
* (3, 2) @ (2, 3) -> (3, 3)
* Note: "@" in Python is the symbol for matrix multiplication.

```python
#Element wise multiplication
print(tensor * tensor)
```

```python
#Matrix Multiplication
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

### Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other(vstack) or side by side(hstack)
* Squeeze - removes all 1 dimensions from a tensor
* Unsqeeze - add a 1 dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted(swapped) in aa certai  way

```python
#create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
#Change the view
z = x.view(1, 9)
z, z.shape
```

```python
#changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other

x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
x_reshaped.shape
```

```python
# Squeezing
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
# Unsqueeze
x_reshaped.unsqueeze(dim=0)
```

```python
# Permute
x_original = torch.rand(size=(224, 224, 3))  # [height, width, color_channel]

x_permuted = x_original.permute(2, 0, 1)  #Shifts axis 0->1, 1->2, 2->0
x_permuted.shape
```

### PyTorch tensors and NumPy

NumPy is a popular scientific python numerical computing library
And because of this, PyTorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> torch.from_numpy(ndarray)
* PyTorch tensor -> NumPy -> torch.Tensor.numpy()

```python
# NumPy array to tensor
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
tensor = tensor + 1
tensor, numpy_tensor
```

### Reproducability (trying to take random out of random)

In short how a neural network learns:

' start with random numbers -> tensor operation -> update random numbers to try and make them better representations of the data -> again -> again -> ... '

To reduce the randomness in nueral networks and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the randomness.

```python
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)

random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
